```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import os as os
```

```python
## criar variavel para armazenar o caminho do deretório com os arquivos

dir_path = 'C:\\Users\\FIC\\Documents\\Eder do Nascimento Bernardo\\arquivos\\003 - Arquivos\\Super Atacadista\\vendas\\'
os.listdir(dir_path)

dir_path
```

```python
## criando uma lista vazia res (repositório) para armazenar os caminhos completos
res = []
```

```python
## lista todos os arquivos que estão dentro de um determinado diretório.

for path in os.listdir(dir_path):
    if os.path.isfile(dir_path + path): ## Verificar se é um arquivo válido
        res.append(dir_path + path)

res
```

```python
## e) Crie um dataframe vazio, e nomeie ele como files_combined.

files_combined = pd.DataFrame()
```

```python
## estrutura para combinar os arquivos

for file in res:
    df1 = pd.read_excel(file)
    files_combined = pd.concat([files_combined,df1])

files_combined
```
